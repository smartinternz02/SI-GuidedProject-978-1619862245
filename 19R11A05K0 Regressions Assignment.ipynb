## Fish Dataset Multi Linear Regression

```python
import pandas as pd
import numpy as np
np
```

```python
ds=pd.read_csv("Fish.csv")
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
x=ds.drop(['Weight'] , axis = 1).values
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer( [("encoder" , OneHotEncoder() ,[0] )] , remainder = "passthrough")
x = ct.fit_transform(x)
```

```python
x
```

```python
y = ds.iloc[:,1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.2 , random_state = 10000 )
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
ypred = mlr.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(ypred,y_test)
```

```python
accuracy
```

```python
yp=mlr.predict(ct.transform( [["Bream" , 24, 25, 30, 11, 4]] ))
```

```python
yp
```
